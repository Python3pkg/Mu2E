# Generate Force Vectors

The goal here is to visualize the force vectors generated by a charged particle moving in a magnetic field.  We'll be looking a much simplier exercise than a full simulation.

```python
from mu2e.dataframeprod import DataFrameMaker
from mu2e import mu2e_ext_path
import matplotlib.pyplot as plt
from mu2e.tools.axes3d_hack import Axes3D
```

```python
 df = DataFrameMaker(mu2e_ext_path+'datafiles/Mau10/Standard_Maps/Mu2e_DSMap', use_pickle=True).data_frame
```

```python
xs = range(-200,1,25)
ys = [100,150]
zs = range(12721, 12900,50)
zs
```

```python
df_trap = df.query('12600<Z<13000 and -800<X<800 and -800<Y<800', inplace=False)
df_trap = df_trap[(df_trap.X.isin(xs)) & (df_trap.Y.isin(ys)) & (df_trap.Z.isin(zs))]
```

```python
df_trap_by = df_trap[['X','Y','Z','Bx','By','Bz']]
```
